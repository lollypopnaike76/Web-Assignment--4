1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url =
https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import requests
import re
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
# Open the Wikipedia website
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
# Find all table rows on the page that contains the dataset details
table = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]')
```

```python
# Create empty lists to store the details
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []

# Iterating through each of the rows of the table
rows = table.find_elements(By.TAG_NAME, 'tr')[1:]

for row in rows:
    columns = row.find_elements(By.TAG_NAME, 'td')
    if len(columns) > 0:
    
        rank = columns[0].text
        name = columns[1].text
        artist = columns[2].text
        upload_date = columns[4].text
        views = columns[3].text

        Rank.append(rank)
        Name.append(name)
        Artist.append(artist)
        Upload_date.append(upload_date)
        Views.append(views)

# Create a DataFrame to store the data and display the result
data = {
    'Rank': Rank,
    'Name': Name,
    'Artist': Artist,
    'Upload Date': Upload_date,
    'Views': Views
}
df = pd.DataFrame(data)

df
```

```python
# Close the browser
driver.quit()
```

7. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls512407256/You
have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes 

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
# Open the imdb.com website
driver.get("https://www.imdb.com/list/ls512407256/")

# Wait for some time to load the fixtures page
time.sleep(3)
```

```python
# Create empty lists to store the details
Name = []
Year_span = []
Genre = []
Run_time = []
Ratings = []
Votes = []

# Scraping the data
series_name= driver.find_elements(By.XPATH,'//div/h3[@class="lister-item-header"]/a')
for series in series_name:
    Name.append(series.text)


try:
    year_span = driver.find_elements(By.XPATH,'//div/h3[@class="lister-item-header"]/span[2]')
    for y in year_span:
        Year_span.append(y.text)
except NoSuchElementException:
    pass


try:
    genre= driver.find_elements(By.CLASS_NAME,'genre')
    for g in genre:
        Genre.append(g.text)
except NoSuchElementException:
    pass


try:
    runtime = driver.find_elements(By.CLASS_NAME,'runtime')
    for p in runtime:
        Run_time.append(p.text)
except NoSuchElementException:
    pass

try:
    rating = driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]')
    for r in rating:
        Ratings.append(r.text)
except NoSuchElementException:
    pass


try:
    votes = driver.find_elements(By.XPATH,'//p[@class="text-muted text-small"][3]')
    for v in votes:
        vote_value = str(v.text)
        if ':' in vote_value:
            vote_value = vote_value.split(':')
        Votes.append(vote_value[1])
except NoSuchElementException:
    pass

```

```python
#displaying the length of items in appended list
print(len(Name),len(Year_span),len(Genre),len(Run_time),len(Ratings),len(Votes))
```

```python
#Creating a dictionary from the extracted data
data = {
    "Names": Name,
    "Year Span": Year_span,
    "Genres": Genre,
    "Run Time": Run_time,
    "Rating": Ratings,
    "Vote": Votes
}
#Creating a DataFrame from the dictionary
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
# Close the WebDriver
driver.quit()
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()
```

```python
# Open the UCI machine learning repositories website
driver.get("https://archive.ics.uci.edu/")

# Wait for some time to load the fixtures page
time.sleep(3)
```

```python
#click the policy tab
policy_click = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/div/div[2]/button')
policy_click.click()
```

```python
# Find and click on the "View All Data Sets" link
view_datasets = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/footer/div[3]/a[2]')
view_datasets.click()
```

```python
# Create empty lists to store the details
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]

#fetching all the required data from all 647 pages
for x in range(0, 600):
    try:
        drop_down = driver.find_elements(By.XPATH, '//div[@class="relative col-span-8 sm:col-span-7"]/*[name()="svg"]')
        for i in drop_down:
            if i is not None:
                i.click()
                time.sleep(1)
        
        dataset_name = driver.find_elements(By.XPATH,'//div[@class="relative col-span-8 sm:col-span-7"]/h2')
        for i in dataset_name:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                Dataset_name.append('-')
            else:
                Dataset_name.append(i.text)

        data_type=driver.find_elements(By.XPATH,'//div/table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[2]')
        for i in data_type:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                Data_type.append('-')
            else:
                Data_type.append(i.text)

        task=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][1]/span')
        for i in task:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                Task.append('-')
            else:
                Task.append(i.text)

        attribute_type=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][2]/span')
        for i in attribute_type:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                Attribute_type.append('-')
            else:
                Attribute_type.append(i.text)

        instances=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][3]/span')
        for i in instances:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                No_of_instances.append('-')
            else:
                No_of_instances.append(i.text)

        attribute=driver.find_elements(By.XPATH,'//div[@class="col-span-3 flex items-center gap-2"][4]/span')
        for i in attribute:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                No_of_attribute.append('-')
            else:
                No_of_attribute.append(i.text)

        year=driver.find_elements(By.XPATH,'//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]')
        for i in year:
            if str(i.text) == ' ' or str(i.text) == None or str(i.text) == 'N/A':
                Year.append('-')
            else:
                Year.append(i.text)


        clicknxt=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
        clicknxt.click()
        time.sleep(1)
        
    except Exception as e:
        break
```

```python
Dataset_name = Dataset_name[:500]
Data_type = Data_type[:500]
Task = Task[:500]
Attribute_type = Attribute_type[:500]
No_of_instances = No_of_instances[:500]
No_of_attribute = No_of_attribute[:500]
Year = Year[:500]
```

```python
#displaying the length of items in appended list
print(len(Dataset_name),len(Data_type),len(Task),len(Attribute_type),len(No_of_instances),len(No_of_attribute),len(Year))
```

```python
# Create a DataFrame to store the data nd display the result
df=pd.DataFrame({
    'Dataset Name':Dataset_name,
    'Data Type':Data_type,
    'Task':Task,
    'Attribute Type':Attribute_type,
    'No of Instances':No_of_instances,
    'No of Attribute':No_of_attribute,
    'Year':Year
})
df
```
